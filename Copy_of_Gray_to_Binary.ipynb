```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

path = '/content/BEACH.jpg'
gray = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

```

```python
mean = np.mean(gray)
thr = int(input("Enter value: "))

small_gray = cv2.resize(gray, (0,0), fx=0.5, fy=0.5)

cv2_imshow(small_gray)

ret1, binary1 = cv2.threshold(gray, mean, 255, cv2.THRESH_BINARY)
ret2, binary2 = cv2.threshold(gray, thr, 255, cv2.THRESH_BINARY)

small_b1 = cv2.resize(binary1, (0,0), fx=0.5, fy=0.5)
small_b2 = cv2.resize(binary2, (0,0), fx=0.5, fy=0.5)

print(mean)
cv2_imshow(small_b1)
cv2_imshow(small_b2)

```
